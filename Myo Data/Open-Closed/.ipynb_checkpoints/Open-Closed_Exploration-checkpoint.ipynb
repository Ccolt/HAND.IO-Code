```python
# import data
closed_full = np.loadtxt('./Close_hand_5x.csv',delimiter=';',usecols=range(8))
open_full = np.loadtxt('./Open_hand_5x.csv', delimiter=';',usecols=range(8))

# create time axes
closed_t = np.linspace(0,len(closed_full)/200, num=len(closed_full))
open_t = np.linspace(0,len(open_full)/200, num=len(open_full))
```

```python
# plot closed-hand data
plt.figure(1)
plt.plot(closed_t[:500],closed_full[:500,0], 'm-')
plt.plot(closed_t[:500],closed_full[:500,1], 'r-')
plt.plot(closed_t[:500],closed_full[:500,2], 'g-')
plt.plot(closed_t[:500],closed_full[:500,3], 'b-')
plt.plot(closed_t[:500],closed_full[:500,4], 'c-')
plt.plot(closed_t[:500],closed_full[:500,5], 'y-')
plt.plot(closed_t[:500],closed_full[:500,6], 'k-')
plt.plot(closed_t[:500],closed_full[:500,7], 'pink')
plt.title("Closed Hand")
plt.xlabel("Time (s)")
plt.ylabel("Raw EMG Signal")

# plot open-hand data
plt.figure(2)
plt.plot(open_t[:500],open_full[:500,0], 'm-')
plt.plot(open_t[:500],open_full[:500,1], 'r-')
plt.plot(open_t[:500],open_full[:500,2], 'g-')
plt.plot(open_t[:500],open_full[:500,3], 'b-')
plt.plot(open_t[:500],open_full[:500,4], 'c-')
plt.plot(open_t[:500],open_full[:500,5], 'y-')
plt.plot(open_t[:500],open_full[:500,6], 'k-')
plt.plot(open_t[:500],open_full[:500,7], 'pink')
plt.title("Open Hand")
plt.xlabel("Time(s)")
plt.ylabel("Raw EMG Signal")  
plt.show()
```
